# Implementing Machine Learning Models

# 1) Logistic Regression

Hyperparameters used for tuning Logistic Regression:-

1)random_state=a random_state parameter may be provided to control the random number generator used.Using an int will produce the same results across different calls. 

2)class_weight=Weights associated with classes in the form {class_label: weight}. If not given, all classes are supposed to have weight one. In this case,we are changing the class_weights in the ratio 1:5,because of imbalanced classes of target variable.

3)solver:-Algorithm to use in the optimization problem. Here,we have used solver as 'sag'.

```python
from time import perf_counter 
t0= perf_counter()

class_weight=dict({0:1,1:2})
from sklearn.linear_model import LogisticRegression

#creating a model
classifier=LogisticRegression(random_state=10,class_weight=class_weight,solver='sag')
classifier.fit(X_train,Y_train)
#---------------------------------------------------------------------------------------------------------------------------
#Prediction
y_pred_logistic=classifier.predict(X_test)
#---------------------------------------------------------------------------------------------------------------------------
#Classification Report
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report,f1_score,recall_score,precision_score

cfm=confusion_matrix(Y_test,y_pred_logistic)
print("CONFUSION MATRIX: ")
print(cfm)

print('Classification Report: ')
print(classification_report(Y_test,y_pred_logistic))

acc1=round(accuracy_score(Y_test,y_pred_logistic),6)
print("Accuracy of the model: %.2f%%" % (acc1*100.0))
acc1=acc1*100.0

f1_score_logistic=round(f1_score(Y_test,y_pred_logistic),4)
print("f1 score of the model: ",f1_score_logistic)

recall_score_logistic=round(recall_score(Y_test,y_pred_logistic),4)
print("recall score of the model: ",recall_score_logistic)

precision_score_logistic=round(precision_score(Y_test,y_pred_logistic),4)
print("precision score of the model: ",precision_score_logistic)

t1 = (perf_counter() - t0)/60
t1=round(t1,2)
print("Time elapsed: ", t1,"minutes") 
```

# Adjusting the threshold-Logistic Regression

In this algorithm, we are appending the values in probability matrix to 0 and 1.

The default threshold is always set to 0.5, such that equal importance is given to both the classes.

But, due to imbalanced classes, we are changing the threshold to 0.41.

```python
from time import perf_counter 
t0= perf_counter()


y_pred_prob=classifier.predict_proba(X_test)
#print(y_pred_prob)
#---------------------------------------------------------------------------------------------------------------------------
for a in np.arange(0.4,0.61,0.01):    #arange(include,exclude,steps)
    predict_mine = np.where(y_pred_prob[:,1] > a, 1, 0)  
    cfm=confusion_matrix(Y_test, predict_mine)
    total_err=cfm[0,1]+cfm[1,0]  #error=type1 error+type2 error
    print("Errors at threshold ", a, ":",total_err, " , type 2 error :",
                cfm[1,0]," , type 1 error:", cfm[0,1])
#---------------------------------------------------------------------------------------------------------------------------
#appending the values in probability matrix to 0 and 1
y_pred_class=[]
for value in y_pred_prob[:,1]:
    if value > 0.41:
        y_pred_class.append(1)
    else:
        y_pred_class.append(0)
#print(y_pred_class)
#---------------------------------------------------------------------------------------------------------------------------
#Classification Report
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report,f1_score

cfm=confusion_matrix(Y_test,y_pred_class)
print("CONFUSION MATRIX: ")
print(cfm)

print('Classification Report: ')
print(classification_report(Y_test,y_pred_class))

acc2=round(accuracy_score(Y_test,y_pred_class),6)
print("Accuracy of the model: %.2f%%" % (acc2*100.0))
acc2=acc2*100.0

f1_score_logistic_tuned=round(f1_score(Y_test,y_pred_class),4)
print("f1 score of the model: ",f1_score_logistic_tuned)

recall_score_logistic_tuned=round(recall_score(Y_test,y_pred_class),4)
print("recall score of the model: ",recall_score_logistic_tuned)

precision_score_logistic_tuned=round(precision_score(Y_test,y_pred_class),4)
print("precision score of the model: ",precision_score_logistic_tuned)

t2 = (perf_counter() - t0)/60
t2=round(t2,2)
print("Time elapsed: ", t2,"minutes") 
```

# Decision Trees

Hyperparameters used for tuning Decision Tree:-

1)criterion-The function to measure the quality of a split. Supported criteria are “gini” for the Gini impurity and “entropy” for the information gain. While using "gini" as criterion, the model performed poorly, so we selected the criterion as "entropy".

2)min_samples_split-The minimum number of samples required to split an internal node.

3)min_samples_leaf-The minimum number of samples required to be at a leaf node. A split point at any depth will only be considered if it leaves at least min_samples_leaf training samples in each of the left and right branches. 

4)max_depth-
The maximum depth of the tree. If None, then nodes are expanded until all leaves are pure or until all leaves contain less than min_samples_split samples.

5)max_leaf_nodes-Grow a tree with max_leaf_nodes in best-first fashion. Best nodes are defined as relative reduction in impurity. If None then unlimited number of leaf nodes.

6)random_state=a random_state parameter may be provided to control the random number generator used.Using an int will produce the same results across different calls. 


```python
from time import perf_counter 
  
t0= perf_counter()

from sklearn.tree import DecisionTreeClassifier

model_decision_tree=DecisionTreeClassifier(criterion='entropy',min_samples_split=5,max_depth=500,
                                           min_samples_leaf=200,max_leaf_nodes=1000,random_state=10)
model_decision_tree.fit(X_train,Y_train)

#---------------------------------------------------------------------------------------------------------------------------
#Prediction
y_pred_dt=model_decision_tree.predict(X_test)

#---------------------------------------------------------------------------------------------------------------------------
#Classification Report
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix,f1_score

cfm=confusion_matrix(Y_test,y_pred_dt)
print("CONFUSION MATRIX: ")
print(cfm)

print('Classification Report: ')
print(classification_report(Y_test,y_pred_dt))

acc3=round(accuracy_score(Y_test,y_pred_dt),6)
print("Accuracy of the model: %.2f%%" % (acc3*100.0))
acc3=acc3*100.0

f1_score_dt=round(f1_score(Y_test,y_pred_dt),4)
print("f1 score of the model: ",f1_score_dt)

recall_score_dt=round(recall_score(Y_test,y_pred_dt),4)
print("recall score of the model: ",recall_score_dt)

precision_score_dt=round(precision_score(Y_test,y_pred_dt),4)
print("precision score of the model: ",precision_score_dt)

t3 = (perf_counter() - t0)/60
t3=round(t3,2)
print("Time elapsed: ",t3,"minutes") 
```

# Voting Based Classifier

Hyperparameters used for tuning Voting Based Classifier:-
1)estimators-We specify the VotingClassifier as a list of (name, sub-estimator) tuples. Fitting the VotingClassifier on the data fits each of the sub-estimators in turn.


```python
from time import perf_counter 
t0=perf_counter()

class_weight=dict({0:1,1:2})
#----------------------------------------------------------------------------------------------------------------------
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier

# create the sub models
estimators = []
model1 = LogisticRegression(random_state=10,class_weight=class_weight)
estimators.append(('log', model1))

model2 = DecisionTreeClassifier(criterion='entropy',min_samples_split=5,max_depth=500,
                                           min_samples_leaf=200,max_leaf_nodes=1000,random_state=10)
estimators.append(('cart', model2))

model3 = RandomForestClassifier(criterion='entropy',min_samples_split=10,max_depth=850,
                                            min_samples_leaf=60,max_leaf_nodes=650,random_state=10)
estimators.append(('random_forest', model3))

#----------------------------------------------------------------------------------------------------------------------

# create the ensemble model
ensemble = VotingClassifier(estimators)
ensemble.fit(X_train,Y_train)
Y_pred_voting=ensemble.predict(X_test)
#print(Y_pred)

#-------------------------------------------------------------------------------------------------------------------------

from sklearn.metrics import classification_report,accuracy_score,confusion_matrix

cfm=confusion_matrix(Y_test,Y_pred_voting)
print("CONFUSION MATRIX: ")
print(cfm)

print('Classification Report: ')

print(classification_report(Y_test,Y_pred_voting))

acc5=round(accuracy_score(Y_test,Y_pred_voting),6)
print("Accuracy of the model: %.2f%%" % (acc5*100.0))
acc5=acc5*100.0

f1_score_voting=round(f1_score(Y_test,Y_pred_voting),4)
print("f1 score of the model: ",f1_score_voting)

recall_score_voting=round(recall_score(Y_test,Y_pred_voting),4)
print("recall score of the model: ",recall_score_voting)

precision_score_voting=round(precision_score(Y_test,Y_pred_voting),4)
print("precision score of the model: ",precision_score_voting)

t5 = (perf_counter()- t0)/60
t5=round(t5,2)
print("Time elapsed: ", t5,"minutes") 
```
